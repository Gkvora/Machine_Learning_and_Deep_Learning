## KNN ( K-Nearest Neighbors )

```python
import pandas as pd
dataset = pd.read_csv("C:\\Users\\A\\Downloads\\Car.csv")
dataset.head(10)
```

```python
from matplotlib import pyplot as plt
plt.scatter(dataset.Age,dataset['Income'])
```

```python
df1 = dataset[dataset.Car==0]
df2 = dataset[dataset.Car==1]
plt.scatter(df1.Age,df1['Income'],color='green')
plt.scatter(df2.Age,df2['Income'],color='red')
```

```python
X = dataset[['Age','Income']]
y = dataset[['Car']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict([[33,149000]]))
```

```python
print(classifier.predict([[25,14900]]))
```

```python
classifier.score(X_test,y_test)
```
